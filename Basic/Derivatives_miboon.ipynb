```
import torch
```

```
x = torch.tensor(2.0, requires_grad = True)
y = 9*x**4 + 2*x**3 + 3*x** 2 + 6*x + 1
y.backward() # 미분
x.grad # 미분 값에 x를 넣어요
```

```
x1 = torch.tensor(1.0, requires_grad = True)
z1 = torch.tensor(2.0, requires_grad = True)

y1 = x1**2 + z1**3
y1.backward() # 미분
x1.grad # x에 대해 미분
```

```
z1.grad # z에 대해 미분
```
