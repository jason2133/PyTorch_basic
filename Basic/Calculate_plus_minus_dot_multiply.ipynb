```
import torch
import matplotlib.pyplot as plt
```

```
t_one = torch.tensor([1, 2, 3])
t_two = torch.tensor([1, 2, 3])
t_one + t_two
```

```
t_one - t_two
```

```
t_one * t_two
```

```
t_one * 5
```

```
dot_product = torch.dot(t_one, t_two) # 행렬 곱셈
print(dot_product)
```

```
torch.linspace(0, 10)
# linspace는 default가 a부터 b까지의 숫자 중 100개를 equally하게 뽑아준다
```

```
torch.linspace(0, 10, 5)
```

```
x = torch.linspace(0, 10, 5)
y = torch.exp(x)
```

```
plt.plot(x.numpy(), y.numpy())
```

```
x1 = torch.linspace(0, 10, 100)
y1 = torch.exp(x1) # exp 함수
plt.plot(x1.numpy(), y1.numpy()) # Numpy
```

```
x2 = torch.linspace(0, 10, 100)
y2 = torch.sin(x2) # Sin 함수
plt.plot(x2.numpy(), y2.numpy())
```
