```
from __future__ import print_function
import torch
```

```
# 무작위로 초기화된 행렬 생성

x = torch.empty(5, 3)
print(x)
```

```
x = torch.zeros(5, 3, dtype=torch.long)
print(x)
```

```
x = torch.tensor([5.5, 3])
print(x)
```

```
x = x.new_ones(5, 3, dtype=torch.double)
print(x)
```

```
x = torch.ones(5, 3, dtype=torch.double)
print(x)
```

```
x = torch.randn_like(x, dtype=torch.float)
print(x)
```

```
print(x.size())
```

### 연산 operations

```
print(x)
```

```
y = torch.rand(5, 3)
print(y)
```

```
print(x+y)
```

```
print(torch.add(x, y))
```

```
result = torch.empty(5, 3)
torch.add(x, y, out=result)
print(result)
```

```
y.add_(x)
print(y)
```

```
print(y[:, 1])
```

```
# 크기 변경 - torch.view

x = torch.randn(4, 4)
y = x.view(16)
z = x.view(-1, 8)

print(x.size(), y.size(), z.size())
```

```
x = torch.randn(1)
print(x)
print(x.item())
```

```
# Torch Tensor -> Numpy
a = torch.ones(5)
print(a)
```

```
b = a.numpy()
print(b)
```

```
a.add_(1)
# add_ 덧셈해서 그걸 바로 적용해라

print(a)
print(b)
```

```
# Numpy -> Torch Tensor
import numpy as np

a = np.ones(5)
b = torch.from_numpy(a)
np.add(a, 1, out=a)

print(a)
print(b)
```

```
# CUDA Tensors

if torch.cuda.is_available():
  device = torch.device("cuda")
  y = torch.ones_like(x, device=device)
  x = x.to(device)
  z = x + y
  print(z)
  print(z.to("cpu", torch.double))
```
