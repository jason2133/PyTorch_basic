```
import torch
```

```
w = torch.tensor(3.0, requires_grad = True) # Weight
b = torch.tensor(1.0, requires_grad = True) # Bias

# y = w*x + b
```

```
def forward(x):
  y = w*x + b
  return y
```

```
x1 = torch.tensor(2)
forward(x1)
```

```
x2 = torch.tensor([[4], [7]]) # 이렇게 넣어도 됨! 대신 [[a], [b]] 형식으로 하나의 [] 안에 들어가도록 만듦.
forward(x2)
```
